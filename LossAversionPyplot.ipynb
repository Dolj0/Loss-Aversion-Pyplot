```
import matplotlib.pyplot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.2, 10, 100)
fig, ax = plt.subplots()

ax.plot([-10,0,20],[-200,0,100])
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_ylabel('U(X)')
ax.set_xlabel('X')
ax.grid(True, which='both')
ax.annotate("Game Won\n U(X) = 50", xy=(10,50), xycoords='data', xytext=(15,15), textcoords='data', arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)
ax.annotate("Game Lost\n U(X) = -200", xy=(-10,-200), xycoords='data', xytext=(-6.5,-180), textcoords='data', arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)
ax.annotate("0,0 See figure 2", xy=(0,0), xycoords='data', xytext=(5,-40), textcoords='data', arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)

plt.show()
```

```
x = np.linspace(0.2, 10, 100)
fig, ax = plt.subplots()

ax.plot([0,0.5,1],[0,1,1])
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_ylabel('v\u0302')
ax.set_xlabel(u'\u03BC')
ax.grid(True, which='both')
ax.annotate("v\u0302=2"+u'\u03BC'+"\ncontractor's \npursuasion payoff", xy=(0.4,0.8), xycoords='data', xytext=(0.6,0.63), textcoords='data', arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)
ax.annotate("Pr("+u'\u03C9'+"=L)=0.2\nv\u0302=0.4", xy=(0.2,0.4), xycoords='data', xytext=(0.4,0.44), textcoords='data', arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)
ax.annotate("(0.5,1)", xy=(0.5,1), xycoords='data', xytext=(0.31,0.93), textcoords='data', arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),)

plt.show()
```
